### Figure 1

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

# enter additional commands here to reproduce the figure below

plt.figure()
plt.plot(a, 'ko-')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
# NOTE!
# - Entering commands and running the cell may cause the figure to disapper
# - If the figure disappears, use Assn-3.html as a reference
```

### Figure 2

Use `data-fig2.npy` for this figure.

```python
b   = np.load( 'data-fig2.npy' )
labels = ['Red', 'Green', 'Blue' ]

plt.figure()
plt.plot(b[:, 0], color='red')
plt.plot(b[:, 1], color='green')
plt.plot(b[:, 2], color='blue')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend(labels)
plt.show()
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c   = np.load( 'data-fig3.npy' )

plt.figure()
plt.scatter(c[:,0], c[:,1] , color='blue', alpha=0.3, s=200)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d   = np.load( 'data-fig4.npy' )
xlabels = ['A', 'B', 'C', 'D', 'E']

plt.figure()
plt.bar(xlabels, d , color='0.7')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
e   = np.load( 'data-fig5.npy' )

plt.figure()
plt.pcolor(e, cmap='RdBu', vmin=-1.0, vmax=1.0)
plt.colorbar()

plt.show()
```
